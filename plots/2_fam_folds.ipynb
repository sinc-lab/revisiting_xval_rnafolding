```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd 

summary = pd.read_csv("../results/2_fam_folds.csv")
summary["family"] = summary.id.str.split("_").str[0]

# boxplot methods (to improve)
plt.figure(figsize=(10, 6))
sns.boxplot(data=summary, x="family", y="f1", hue="method")
plt.xticks(rotation=45)
plt.xlabel("Family")
plt.ylabel("F1 score")
plt.title("Fam-fold")
plt.tight_layout()
```

```python
summary[summary.method.isin(["sincFold", "UFold", "REDfold"])].groupby(["method"]).f1.mean()
```
