```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd 

partition_name = "3_clustering_folds"
summary = pd.read_csv(f"../results/{partition_name}.csv")

plt.figure(figsize=(10, 6))
sns.boxplot(data=summary, y="f1", x="method")
plt.xticks(rotation=45)
plt.xlabel("Method")
plt.ylabel("F1 score")
plt.title(partition_name)
plt.tight_layout()
```

```python
summary[summary.method.isin(["sincFold", "UFold", "REDfold"])].groupby(["method"]).f1.mean()
```
