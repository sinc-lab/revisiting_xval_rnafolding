```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd 

partition_name = "5_hl_folds"
summary = pd.read_csv(f"../results/{partition_name}.csv")

# boxplot methods (to improve)
plt.figure(figsize=(10, 6))
sns.boxplot(data=summary, y="f1", x="fold", hue="method")
plt.xticks(rotation=45)
plt.xlabel("Fold")
plt.ylabel("F1 score")
plt.title("HL folds")
plt.tight_layout()
```

```python
summary[summary.method.isin(["sincFold", "UFold", "REDfold"])].groupby(["method"]).f1.mean()
```
